## 4. Test Matplotlib Visualization
Testing plotting capabilities essential for data visualization and result analysis:

```python
if 'plt' in globals() and 'np' in globals():
    print("üìä Testing Matplotlib Visualization:")
    
    # Set up the plotting area
    plt.figure(figsize=(12, 8))
    
    # Create sample data
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    
    # Subplot 1: Line plot
    plt.subplot(2, 2, 1)
    plt.plot(x, y1, 'b-', label='sin(x)', linewidth=2)
    plt.plot(x, y2, 'r--', label='cos(x)', linewidth=2)
    plt.title('Trigonometric Functions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Subplot 2: Scatter plot
    plt.subplot(2, 2, 2)
    x_scatter = np.random.randn(50)
    y_scatter = 2 * x_scatter + np.random.randn(50)
    plt.scatter(x_scatter, y_scatter, alpha=0.6, c='purple')
    plt.title('Random Scatter Plot')
    plt.xlabel('X values')
    plt.ylabel('Y values')
    plt.grid(True, alpha=0.3)
    
    # Subplot 3: Histogram
    plt.subplot(2, 2, 3)
    data = np.random.normal(0, 1, 1000)
    plt.hist(data, bins=30, alpha=0.7, color='green', edgecolor='black')
    plt.title('Normal Distribution Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)
    
    # Subplot 4: Simple bar chart
    plt.subplot(2, 2, 4)
    categories = ['Training', 'Validation', 'Testing']
    values = [0.85, 0.78, 0.82]
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
    plt.bar(categories, values, color=colors, alpha=0.8)
    plt.title('Model Performance')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1)
    plt.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.suptitle('ü¶∑ MAHT-Net Visualization Test', fontsize=16, y=1.02)
    plt.show()
    
    print("‚úÖ Matplotlib is working correctly!")
    print(f"   Version: {plt.matplotlib.__version__}")
    
else:
    print("‚ùå Matplotlib or NumPy not available - plotting functionality disabled")
```
